```python
import pandas as pd
```

```python
df2019 = pd.read_csv('201908-citibike-tripdata.csv')
df2019.head()
```

```python
dfJC2019 = pd.read_csv('JC-201908-citibike-tripdata.csv')
dfJC2019.head()
```

```python
df201908 = pd.concat([df2019, dfJC2019])
df201908.head()
```

```python
df201908.count()
```

```python
df201908 = df201908.rename(columns={"tripduration": "Trip Duration (sec)", 
                    "starttime": "Start Time", 
                   "stoptime": "Stop Time",
                   "start station id": "Start Station ID",
                   "start station name": "Start Station Name",
                   "start station latitude": "Start Station Latitude",
                   "start station longitude": "Start Station Longitude",
                   "end station id": "End Station ID",
                   "end station name": "End Station Name",
                   "end station latitude": "End Station Latitude",
                   "end station longitude": "End Station Longitude",
                   "bikeid": "Bike ID",
                   "usertype": "User Type",
                   "birth year": "Birth Year",
                   "gender":"Gender"})
df201908.head()
```

```python
#exclude ages: identify the age 2 standard deviations lower than the mean. After calculating this number, mean - 2stdev, remove the tail end of the data
df201908 = df201908.drop(df201908.index[(df201908['Birth Year'] < df201908['Birth Year'].mean()-(2*df201908['Birth Year'].std()))])
```

```python
df201908.count()
```

```python
#Add Age column
df201908["Age"] = 2019 - df201908["Birth Year"]
df201908.head()
```

```python
#drop circular trips
df201908 = df201908.drop(df201908.index[(df201908['Start Station Latitude']-df201908['End Station Latitude'] == 0)])
```

```python
df201908.count()
```

```python
#drop n/a
df201908 = df201908.dropna(how='any').reset_index(drop=True)
df201908.count()
```

```python
df201908.to_csv('citibike_201908.csv', index=False)
```
